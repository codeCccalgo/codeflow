```python
import random 
import math 
import os
import subprocess
import shutil
import make as mk 
def add_freopen_to_cpp(cpp_file):
    # 读取现有的 C++ 文件内容
    with open(cpp_file, "r") as file:
        lines = file.readlines()

    # 要插入的 freopen 和 fclose 代码
    freopen_code = [
        '    freopen("wk1.in", "r", stdin);  \n',
        '    freopen("wk2.out", "w", stdout);  \n'
    ]
    fclose_code = [
        '    fclose(stdin);\n',
        '    fclose(stdout);\n'
    ]

    # 查找 main 函数的位置
    in_main = False
    modified_lines = []
    for line in lines:
        modified_lines.append(line)
        if "main" in line:
            modified_lines.extend(freopen_code)
            # in_main = False
        elif "return 0;" in line:
            # 在 return 0; 前插入 fclose 代码
            # print("check")
            modified_lines = modified_lines[:-1]  # 删除 return 0;
            modified_lines.extend(fclose_code)
            modified_lines.append('    return 0;\n')

    # 将修改后的代码写回文件
    with open(cpp_file, "w") as file:
        file.writelines(modified_lines)

# 示例用法
cpp_file = "cod.cpp"
add_freopen_to_cpp(cpp_file)
```

```python
import os
import zipfile

# 指定要压缩的目录和目标压缩文件
directory_to_zip = 'data'  # 要压缩的目录
zip_file_name = 'use.zip'  # 生成的压缩文件名

def zip_directory(directory, zip_filename):
    # 创建一个新的 ZIP 文件
    with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
        # 遍历目录中的所有文件和子目录
        for root, dirs, files in os.walk(directory):
            for file in files:
                # 构造文件的完整路径
                file_path = os.path.join(root, file)
                # 相对路径相对于目录，防止在 ZIP 文件中有绝对路径
                relative_path = os.path.relpath(file_path, directory)
                # 将文件添加到 ZIP 文件中
                zipf.write(file_path, relative_path)

    print(f'Created {zip_filename} successfully.')

# 调用函数创建 ZIP 文件
zip_directory(directory_to_zip, zip_file_name)
```
